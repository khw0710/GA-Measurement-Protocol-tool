```
from google_measurement_protocol import event, report
import csv

with open('/content/social_type_output.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

data
```

```
for row in range (1,len(data)):
  eventpackage = event('ml-seg Project','Social Type', label = data[row][1])
  report('UA-39179668-1', data[row][0], eventpackage,  ua='Opera/9.80 (Windows NT 6.0) Presto/2.12.388 Version/12.14', ni='1')
```
